```python
#Parse logs, decorate with some extra features
logs = []
numeric_fields = ['Price', 'MarketSize', 'Size']
with open("test.log", "rb") as f:
    raw_logs = f.readlines()
    raw_logs = [log.split("\t") for log in raw_logs]

for log in raw_logs:
    ts = log[0]
    source = log[1]
    log = {item.split(":")[0]:item.split(":")[1].strip() for item in log[2:]}
    log["Timestamp"] = datetime.strptime(ts, "%H:%M:%S.%f")
    log["Source"] = source      
    if source == "SE":
        for field in numeric_fields:
            log[field] = float(log[field])
    logs.append(log)

#Adding some additional features 
for index, log in enumerate(logs):
    if log["Source"] == "SE":
        #Time in microseconds
        log["Tick_to_OM"] = (logs[index+1]["Timestamp"] - log["Timestamp"]).total_seconds() * 10**6
        log["Notional_size"] = log["Size"] * log["Price"]
        orderid = logs[index+1]["OrderId"]
        c = 1
        while c < len(logs):
            if logs[index + c]["Source"] == "OM" and logs[index + c]["OrderId"] == orderid:
                if logs[index + c]["OrderEventType"] in ["FILL","CANCEL","REJECT"]:
                    log["Outcome"] = logs[index + c]["OrderEventType"]
                    c = 0 
                    break
            c += 1
        log["IsFill"] = 0 if log["Outcome"] != "FILL" else 1

se = pd.DataFrame([log for log in logs if log["Source"]=="SE"])
om = pd.DataFrame([log for log in logs if log["Source"]=="OM"])
```

```python
#Train a random forest classifier to get a broader look at feature importance 

#The rejects never reach market, so i'm restricting this model only orders that appear in the orderbook 
se_train = se[se.Outcome != "REJECT"]
categorical_variables = list(set(se_train.select_dtypes(include=["object"]).columns) - set(["Strike","OptionId","Source","Outcome"]))
numeric_variables = list(set(se_train.select_dtypes(exclude=["object"]).columns) - set(["Timestamp","IsFill"])) 
dummy_train = pd.get_dummies(se_train[categorical_variables])
xtrain = pd.concat([se_train[numeric_variables], dummy_train], axis=1)
ytrain = se_train["IsFill"]

rfc = RandomForestClassifier(n_estimators=300, n_jobs=-1)
rfc.fit(xtrain, ytrain)
```

```python
#Arrange and order features with their importance
columns = xtrain.columns
feature_importance = {columns[index]:importance for index, importance in enumerate(rfc.feature_importances_)}
sorted_importance = sorted(feature_importance.items(), key=operator.itemgetter(1))
values = [i[1] for i in sorted_importance]
keys = [i[0] for i in sorted_importance]
```

```python
"""
Based on our random forest classifier, the features most predictive of fills are the notional size 
of the order placed, the triggering MarketSize, and the time from place to acknowledgement by the OM 

From the EDA earlier, the trend seems to be that smaller notional orders are more likely to get filled.
Also, this plot confirms earlier observations that orders on PHLX get more fills, along with shorter
dated options.
"""
# Plot the feature importances of the forest
plt.figure(figsize=(15,8))
plt.bar(range(len(feature_importance)), values, align='center')
plt.xticks(range(len(feature_importance)), keys)
plt.title("Feature Importance")
plt.xticks(rotation=90)
plt.show()
```
